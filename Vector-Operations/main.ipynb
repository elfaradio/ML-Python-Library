Plotting A Vector


```python

u = [4, 5, 6]
v = [1, 2, 3]

plt.quiver(0, 0, 5, 7, scale_units='xy', angles='xy', scale=1, color='r')
plt.quiver(0, 0, -5, 7, scale_units='xy', angles='xy', scale=1, color='b')
plt.grid(True)
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.show()
```
