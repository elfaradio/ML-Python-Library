```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as ns
```

```python
data = pd.read_csv('Placement_Dataset.csv')
```

```python
data.shape
data.isnull().sum()
```

Central Tendencies:
Mean,median,mode

```python
''' ðŸ“Š What is Skewed Distribution?

In statistics, skewness tells us how much a distribution deviates from being symmetric (like a bell-shaped normal distribution).

Normal distribution (no skew): perfectly symmetric (mean = median = mode).

Skewed distribution: one tail is longer than the other.

ðŸ“Œ Types of Skewness

Positive Skew (Right Skew)

Tail is stretched to the right (higher values).

Mean > Median > Mode.

Example: Income distribution (few very rich people).

Negative Skew (Left Skew)

Tail is stretched to the left (lower values).

Mean < Median < Mode.

Example: Age at retirement (most retire around 60 65, but a few much earlier).'''

fig, ax = plt.subplots(figsize=(8, 8))
ns.histplot(data["salary"], kde=True, bins=30, ax=ax)

ax.set_title("Distribution of Salary", fontsize=16)
ax.set_xlabel("Salary")
ax.set_ylabel("Frequency")

plt.show()
```

As this is skew type so we use median

```python
'''ðŸ› ï¸ Ways to Handle Missing Values
1. Identify Missing Values

In Pandas:
import pandas as pd

# Check how many missing values in each column
print(dataset.isnull().sum())
2. Drop Missing Values

If the dataset is large and missing values are very few, you can drop them.
dataset = dataset.dropna()           # drop rows with missing values
dataset = dataset.dropna(axis=1)     # drop entire columns with missing values
3. Fill (Impute) Missing Values

ðŸ‘‰ Useful when you donâ€™t want to lose data.

(a) Fill with a Constant
dataset["salary"].fillna(0, inplace=True)       # fill with 0
dataset["department"].fillna("Unknown", inplace=True)  # fill with string
dataset["salary"].fillna(dataset["salary"].mean(), inplace=True)
dataset["salary"].fillna(dataset["salary"].median(), inplace=True)
dataset["department"].fillna(dataset["department"].mode()[0], inplace=True)
dataset.fillna(method="ffill", inplace=True)  # fill with previous value
dataset.fillna(method="bfill", inplace=True)  # fill with next value
'''


data['salary'].fillna(data['salary'].median(), inplace=True)
```
