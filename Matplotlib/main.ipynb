```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# create a customized line plot

plt.plot(x, y, color='red', linestyle='--',
         marker='o', linewidth=3, markersize=9)
plt.grid(True)
```
